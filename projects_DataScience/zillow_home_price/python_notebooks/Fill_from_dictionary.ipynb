The purpose of this notebook is to create a function that takes as inputs a dataframe and a dictonary and fills in the numeric features as classes from the dictionary

```python
import pandas as pd
import xlrd
```

```python
def fill_from_dictionary(path_to_data, path_to_dictionary):
    data = pd.read_csv(path_to_data)
    
    #Lower all column headders in data
    columns_lowered = []
    for col in data.columns:
        columns_lowered.append(col.lower())

    data.columns = columns_lowered
    
    #Read in data dictionary
    xl = pd.ExcelFile(path_to_dictionary)
    
    #make dictionary of each sheet from dictionary excel file
    d= {}
    for x in xl.sheet_names[1:]:
        d["{0}".format(x).lower()] = pd.read_excel(path_to_dictionary, sheetname= x, index_col=0)
    
    
    for key in d:
        d2 = d[key].to_dict()
        for key2 in d2:
            data[key].replace(d2[key2], inplace = True)
            data[key].fillna('unknown', inplace = True)
    return(data)
```

```python
data = fill_from_dictionary('../raw_data/properties_2016.csv', '../raw_data/zillow_data_dictionary.xlsx')
```
